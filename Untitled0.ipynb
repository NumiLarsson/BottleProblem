```python
from sklearn.model_selection import train_test_split

targetColumnName = "6"

features_numpy = df.loc[:,df.columns != targetColumnName].values/255 # Normalize pixels to [0,1]
targets_numpy = df[[targetColumnName]].values

features_train, features_test, targets_train, targets_test = train_test_split(features_numpy, targets_numpy, test_size=0.2, random_state=42)
```

```python
# Show values normalized to values between 0 and 1
features_numpy
```

```python
device = (
    "cuda"
    if torch.cuda.is_available()
    else "mps"
    if torch.backends.mps.is_available()
    else "cpu"
)
print(f"Using {device} device")
```

```python
from torch import nn

class NeuralNetwork(nn.Module):
    def __init__(self):
        super().__init__()
        self.flatten = nn.Flatten()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(28*28, 512),
            nn.ReLU(),
            nn.Linear(512, 512),
            nn.ReLU(),
            nn.Linear(512, 512),
            nn.ReLU(),
            nn.Linear(512, 10),
        )

    def forward(self, x):
        x = self.flatten(x)
        logits = self.linear_relu_stack(x)
        return logits
```

```python
net = NeuralNetwork()

model = net.to(device)
print(model)
```

```python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(net.parameters(), lr=0.001, momentum=0.9) # SGD (stochastic gradient descent)
```

```python
#transform = transforms.Compose(
#    [transforms.ToTensor(),
#     transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))])

batch_size = 4

# create feature and targets tensor for train set.
features_train_torch = torch.from_numpy(features_train)
targets_train_torch = torch.from_numpy(targets_train.flatten()).type(torch.LongTensor) # data type is long

# create feature and targets tensor for test set.
features_test_torch = torch.from_numpy(features_test)
targets_test_torch = torch.from_numpy(targets_test.flatten()).type(torch.LongTensor) # data type is long

# batch_size
batch_size = 100

# Pytorch train and test sets
train_dataset = torch.utils.data.TensorDataset(features_train_torch, targets_train_torch)
test_dataset = torch.utils.data.TensorDataset(features_test_torch, targets_test_torch)

# data loader
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size = batch_size, shuffle = False)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size = batch_size, shuffle = False)

epochsToRun = 30
train_loss = []
test_loss = []

for epoch in range(epochsToRun):  # loop over the dataset multiple times

    running_loss = 0.0
    for i, data in enumerate(train_loader, 0):

        # get the inputs; data is a list of [inputs, labels] or [X, y]
        X, y = data # X = batch_size bilder => 100 bilder; y = facit

        # zero the parameter gradients
        optimizer.zero_grad()

        # forward + backward + optimize
        y_pred = net(X)
        loss = criterion(y_pred, y)
        loss.backward()
        optimizer.step()

        # print statistics
        running_loss += loss.item()
        if i % 10 == 9:    # print every 10 batches
            print(f'[{epoch + 1}, {i + 1:5d}] loss: {running_loss / 10:.3f}')
            running_loss = 0.0
    y_train_pred = net(features_train_torch)
    y_test_pred = net(features_test_torch)
    loss_train = criterion(y_train_pred, targets_train_torch)

    loss_test = criterion(y_test_pred, targets_test_torch)

    train_loss.append(loss_train.item())
    test_loss.append(loss_test.item())

print('Finished Training')
```

```python
import matplotlib.pyplot as plt

plt.plot(range(1, epochsToRun+1), train_loss, label='train')
plt.plot(range(1, epochsToRun+1), test_loss, label='tests')
plt.legend()
plt.show()
```

```python
def evaluate_multiclass_classification_model(dataset, model, number_of_classes):

    correct_per_class = np.zeros(number_of_classes)

    for X, y in dataset:

        pred = model(X[None, :])
        correct = (pred.argmax(1) == y).type(torch.float).sum().item() #Is 1 if prediction matches label, y, otherwise 0
        index = int(y.item())
        correct_per_class[index] = correct_per_class[index] + correct

    return correct_per_class
```

```python
numbersOfNumbersClasses = 10
correct_per_class = evaluate_multiclass_classification_model(test_dataset, net, numbersOfNumbersClasses)
classes, total_per_class = np.unique(targets_test, return_counts = True)
accuracy_per_class = (correct_per_class * 1.0) / total_per_class

print(correct_per_class)
print(classes)
print(total_per_class)
print(accuracy_per_class)
```
